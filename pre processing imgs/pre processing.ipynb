# Load all the stuff I need.

```python
import cv2
import os
import random
import matplotlib.pylab as plt
from glob import glob
import pandas as pd
import numpy as np
```

# Path to the Images

```python
# ../input/
PATH = os.path.abspath(os.path.join('..', 'ML_autoencoder'))

# ../input/sample/images/
pp_tt_dir = os.path.join(PATH, "colors")
pp_jj_dir = os.path.join(PATH, "colors_ppjj")

# ../input/sample/images/*.png
images_tt = glob(os.path.join(pp_tt_dir, "*.png"))
images_jj = glob(os.path.join(pp_jj_dir, "*.png"))
```

# Check Images

```python
r = random.sample(images_tt, 3)
plt.figure(figsize=(16,16))
plt.subplot(131)
plt.imshow(cv2.imread(r[0]))
print(r[0])
print(r[1])
print(r[2])

plt.subplot(132)
plt.imshow(cv2.imread(r[1]))

plt.subplot(133)
plt.imshow(cv2.imread(r[2]))
plt.show()
```

# Create a function to read the Images and convert to numpy arrays

```python
def proc_images():
    x0 = [] # images as arrays
    y0 = [] # labels Infiltration or Not_infiltration
    x1 = []
    y1 = []
    for img1 in images_tt:        
        full_size_image1 = cv2.imread(img1)
        x1.append(full_size_image1)
        y1.append(1)
    for img0 in images_jj:
        full_size_image0 = cv2.imread(img0)
        x0.append(full_size_image0)
        y0.append(0)
        
    return x1, y1, x0, y0
```

```python
x_tt,y_tt,x_jj,y_jj = proc_images()
```

```python
x_full = np.concatenate([x_tt,x_jj])
y_full = np.concatenate([y_tt,y_jj])
```
